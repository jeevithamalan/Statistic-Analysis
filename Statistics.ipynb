```python
import numpy as np
from scipy.stats import ttest_1samp
from scipy import stats
```

```python
sample=df.total_sulfur_dioxide.sample(frac=0.05)
sample
```

```python
t_test,p_value=stats.ttest_1samp(sample,df.total_sulfur_dioxide.mean())
print(f'T_TEST:{t_test} \n P_VALUE:{p_value}')
```

```python
if p_value<0.05:
    print(' Null Hypothesis H0 is rejected,Ha is accepted,independentdata not distributed properly')
else:
    print('Null Hypothesis H0 is accepted,Ha is rejected,dependent,data is distributed properly')
```

```python
for i in range(30):
    sample=df.total_sulfur_dioxide.sample(frac=0.04)
    t_test,p_value=stats.ttest_1samp(sample,df.total_sulfur_dioxide.mean())
    print(f'T_TEST:{t_test} \n P_VALUE:{p_value}')
    if p_value<0.05:
        print('Null Hypothesis H0 is rejected,Ha is accepted,independentdata not distributed properly')
    else:
        print('Null Hypothesis H0 is accepted,Ha is rejected,dependent,data is distributed properly')
```

```python
import pandas as pd
import statsmodels.api as sm
```

```python
df=pd.read_excel("/content/Wine_Quality_Data.xlsx")
df
```

```python
group=pd.unique(df['quality'].values)
group
```

```python
data={}
for i in group:
  data[i]=df[df['quality']==i]
data
```

```python

group_data = {value: df[df['quality'] == value]['quality'].values for value in group}

# Perform ANOVA
f_value, p_value = stats.f_oneway(*group_data.values())

print(f"F-value: {f_value}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("H0 - Null hypothesis is rejected: There is a significant difference among groups.")
else:
    print("Ha - Null hypothesis is not rejected: There is no significant difference among groups.")
```
